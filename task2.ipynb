```python
def largest(a,b,c):
    l=a
    if b>l:
        l=b 
    if c>l: 
        l=c
    return l

print(largest(25,35,15))
```

```python
def str_len(str):
    count=0
    for i in str:
        count+=1
    return count

print(str_len("Hiii this is Srikari"))
```

```python
def prime(n):
    if n<=1:
        print("Not a prime number")
        return 
    for i in range(2,n//2+1):
        if n%i==0:
            print("Not a prime number")
            return
    print("Prime number")   

    

prime(20)
prime(19)
```

```python
def fib(n):
    if n<=0:
        return "invalid"
    elif n==1:
        return 0    
    elif n==2:
        return 1
    else:
        return fib(n-1)+fib(n-2)    

fib(10)
```

```python
def rev_str(str):
    l=len(str)
    if l==0:
        return str
    else:
        return rev_str(str[1:])+str[0]

rev_str("Hello world")
```
